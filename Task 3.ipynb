```python
plt.scatter(df['Sales'], df['Profit'])
```

```python
region_sale = df.groupby(by='Region').sum()['Sales']
region_sale = region_sale.sort_values(ascending=False)
region_sale.plot(kind='bar')
plt.title('Regionwise Sales', fontsize=15)
```

```python
state_sale = df.groupby(by='State').sum()['Sales']
state_sale = state_sale.sort_values(ascending=False)
plt.figure(figsize=(16,8))
state_sale.plot(kind='bar')
plt.title('Statewise Sales', fontsize=15)
```

```python
category_sale = df.groupby(by='Category').sum()['Sales']
category_sale = category_sale.sort_values(ascending=False)
category_sale.plot(kind='bar')
plt.title('Categorywise Sales', fontsize=15)
```

```python
sub_category_sale = df.groupby(by='Sub-Category').sum()['Sales']
sub_category_sale = sub_category_sale.sort_values(ascending=False)
plt.figure(figsize=(12,6))
sub_category_sale.plot(kind='bar')
plt.title('Sub-Categorywise Sales', fontsize=15)
```

```python
sub_category_quantity = df.groupby(by='Sub-Category').sum()['Quantity']
sub_category_quantity = sub_category_quantity.sort_values(ascending=False)
plt.figure(figsize=(12,6))
sub_category_quantity.plot(kind='bar')
plt.title('Sub-Categorywise Quantity', fontsize=15)
```

```python
state_profit = df.groupby(by='State').sum()['Profit']
state_profit = state_profit.sort_values(ascending=False)
plt.figure(figsize=(16,8))
state_profit.plot(kind='bar')
plt.title('Statewise Profit', fontsize=15)
```

```python
region_profit = df.groupby(by='Region').sum()['Profit']
region_profit = region_profit.sort_values(ascending=False)
region_profit.plot(kind='bar')
plt.title('Regionwise Profit', fontsize=15)
```

```python
sub_category_profit = df.groupby(by='Sub-Category').sum()['Profit']
sub_category_profit = sub_category_profit.sort_values(ascending=False)
plt.figure(figsize=(10,6))
sub_category_profit.plot(kind='bar')
plt.title('Sub-categorywise Profit', fontsize=15)
```

```python
state_sales_profit = df.groupby(by='State').sum()[['Profit','Sales']]
state_sales_profit.plot.bar(title='Statewise Sales and Profit', figsize=(24,10))
plt.title('Statewise Sales and Profit', fontsize=25)
```

```python
sub_category_sales_profit = df.groupby(by='Sub-Category').sum()[['Profit','Sales']]
sub_category_sales_profit.plot.bar(figsize=(16,6))
plt.title('Sub-categorywise Sales and Profit', fontsize=15)
```

```python
sub_category_quantity_profit = df.groupby(by='Sub-Category').sum()[['Profit','Quantity']]
sub_category_quantity_profit.plot.bar(figsize=(16,10))
plt.title('Sub-categorywise Quantity and Profit', fontsize=15)
```

```python
df['Discount'].value_counts()
```

```python
df.groupby(by='Discount').sum()[['Sales', 'Profit']].plot(figsize=(10,6))
plt.ylabel('Sales')
plt.title('Discount vs Sales and Profit')
```
